### 使用决策树预测隐形眼镜类型

```python
%matplotlib inline
import trees
import tree_plotter
```

```python
# 数据每一行为一个样本，样本中数据用tab隔开，最后一项为类别，全部数据都是标签型数据
# 数据准备
filename = './data/lenses.txt'
lenses = []
with open(filename) as f:
    for inst in f:
        lenses.append(inst.strip().split('\t'))
lenses_labels = ['age','prescript','astigmatic','tearRate']
```

```python
lenses_tree = trees.create_tree(lenses,lenses_labels)
print(lenses_tree)
tree_plotter.create_plot(lenses_tree)
```
