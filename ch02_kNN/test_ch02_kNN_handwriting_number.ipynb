```python
# 测试读取数据
def test_input_data():
    filename = "./data/digits/testDigits/0_1.txt"
    test_vector = kNN.img2vector(filename)
    print(test_vector[0,0:31]) # 输出前32位
    print(test_vector[0,32:63]) # 输出33~64位
test_input_data()
```

```python
training_digits_filename = './data/digits/trainingDigits'
test_digits_filename = './data/digits/testDigits'
# 数据文件命名格式：数字类别_该数字的第几个样本.txt
def handwriting_class_test():
    # 获取训练数据
    hw_labels = []
    training_file_list = os.listdir(training_digits_filename)  # 获取训练样本的路径列表
    m = len(training_file_list)
    training_mat = np.zeros((m,1024))  # 申请空间，m行每一行长度为1024
    for i in range(m):
        file_name_str = training_file_list[i]
        file_str = file_name_str.split('.')[0]
        class_num_str = int(file_str.split('_')[0])  # 获取样本的类别
        hw_labels.append(class_num_str)
        # 将第i个样本的数据转化为一个向量复制到训练矩阵的第i行
        training_mat[i,:] = kNN.img2vector(os.path.join(training_digits_filename,file_name_str))
    # 获取测试数据并开始测试
    test_file_list = os.listdir(test_digits_filename)
    error_count = 0.0
    m_test = len(test_file_list)
    for i in range(m_test):
        file_name_str = test_file_list[i]
        file_str = file_name_str.split('.')[0]
        class_num_str = int(file_str.split('_')[0])
        vector_under_test = kNN.img2vector(os.path.join(test_digits_filename,file_name_str))
        classifier_result = kNN.classify0(vector_under_test,training_mat,hw_labels,3)
        print("the classifier came back with: %d, the real answer is: %d" % (classifier_result,class_num_str))
        if classifier_result != class_num_str:
            error_count += 1.0
    print("")
    print("the total number of errors is: %d" % (error_count,))
    print("the total error rate is: %f" %(error_count/float(m_test),))

handwriting_class_test()
```
